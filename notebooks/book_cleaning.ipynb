```python
def sort_dates(date_list):
    month_key = {
        "January": 1,
        "February": 2,
        "March": 3,
        "April": 4,
        "May": 5,
        "June": 6,
        "July": 7,
        "August": 8,
        "September": 9,
        "October": 10,
        "November": 11,
        "December": 12
    }
    
    year_key = {
        "2011": 10,
        "2012": 50,
        "2013": 100,
        "2014": 200,
        "2015": 300,
        "2016": 400,
        "2017": 500,
        "2018": 600,
        "2019": 700,
        "2020": 800,
        "2021": 900
    }
    
    return [year_key[date.split()[1]] + month_key[date.split()[0]] for date in date_list]
        
```

```python
plt.close("all")
data_over_time = books[["title", "finished"]].copy()
data_over_time["year_read"] = data_over_time["finished"].dt.year
data_over_time["month_read"] = data_over_time["finished"].dt.month_name()
data_over_time.dropna(inplace=True)


data_over_time["timeline"] = (
    data_over_time["month_read"]
    + " " +data_over_time["year_read"].astype("int").astype("str")
)

missing_months = pd.DataFrame(
    {"timeline": ["August 2019", "September 2019", "November 2019", "November 2017"], 
     "title": [0, 0, 0, 0],
     "key": [208, 209, 211, 511]
    }
)

books_over_time = data_over_time[data_over_time["finished"].dt.year >= 2015].groupby("timeline")["title"].count().reset_index()
books_over_time = pd.concat([books_over_time, missing_months])
books_over_time["key"] = sort_dates(books_over_time["timeline"])
books_over_time.sort_values(by="key", inplace=True)

timeline = books_over_time["timeline"]

x_values = [label if label.startswith("January") or label.startswith("June") else "" for label in timeline]

plt.figure(figsize=(10, 6))
ax = plt.subplot()
ax.set_xticks(range(len(x_values)))
ax.set_xticklabels(x_values, rotation=90)
ax.set_yticks(range(18))
plt.plot(timeline, books_over_time["title"], marker="o")
plt.title("Number of Books Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Books")
plt.axhline(
    np.mean(books_over_time["title"]), 
    label="Average books in a month: " + str(np.ceil(np.mean(books_over_time["title"]))), 
    color="orange"
)

# plt.axvline("March 2020", label = "Lockdowns in effect", color="green", linestyle="--")
plt.axvline("October 2018", label="Move to Northern California", color="red", linestyle="--")
plt.axvline("August 2020", label="Move to Missouri", color="orange", linestyle="--")
plt.axvline("June 2017", label="Graduated from UCI", color="violet", linestyle="--")
plt.legend(loc=3)
plt.show()
```
