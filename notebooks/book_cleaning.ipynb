# Cleaning

```python
authors = books["author"]
```

```python
authors
```

```python
split_names = authors.str.split(";", n=1, expand=True)
```

```python
split_names_again = split_names[0].str.split(",", expand=True)
```

```python
split_names_again.columns = ["fname", "lname"]
```

```python
split_names_again["full_name"] = split_names_again.apply(lambda row: row["lname"] + " " + row["fname"] if not pd.isnull(row["lname"]) else row["fname"], axis=1)
```

```python
split_names_again
```

```python
books["author"] = split_names_again["full_name"]
```

```python
split_names.columns = ["authors1", "other_authors"]
```

```python
other_authors_split = split_names["other_authors"].str.split(",", expand=True)
```

```python
other_authors_split.columns = ["name1", "name2", "name3"]
```

```python
other_authors_split
```

```python
other_authors_split["full_names"] = other_authors_split.apply(lambda row: row["name2"] + " " + row["name1"] if not (pd.isnull(row["name1"]) or pd.isnull(row["name2"])) and pd.isnull(row["name3"]) else pd.NA, axis=1)
```

```python
other_authors_split.loc[
    other_authors_split["name2"] == " Ola; Ronnlund", 
    "full_names"
] = "Ola Rosling; Anna Rosling Ronnlund"


```

```python
other_authors_split
```

```python
books["other_authors"] = other_authors_split["full_names"]
```

```python
books.loc[books["title"] == "How I Learned to Understand the World", "other_authors"] = "Fanny Hargestam"
```

```python
books["format"] = books["format"].str.lower()
books["genre"] = books["genre"].str.lower()
books["f_nf"] = books["f_nf"].str.lower()
books["dnf"] = books["dnf"].str.lower()
books["reread"] = books["reread"].str.lower()
books["price"] = books["price"].str.replace("$", "", regex=False)
```

```python
books.dtypes
```

```python
hours_minutes = books["duration"].str.split(":", expand=True)
```

```python
hours_minutes.columns = ["hours", "minutes", "seconds"]
```

```python
hours_minutes["total_duration"] = (pd.to_numeric(hours_minutes["hours"]) * 60) + pd.to_numeric(hours_minutes["minutes"])
```

```python
books["duration"] = hours_minutes["total_duration"]
```

```python
for col in books.columns:
    try:
        books[col] = books[col].str.strip()
    except:
        print(col)
# Should recast start/finish to datetime if that step is done before this one.
```

```python
books["duration"]
```

```python
books[books["imprint"] == "St. Martin's Press"]["publishing_house"]
```

```python
books["imprint"] = books["imprint"].str.replace("W.W.", "W. W.", regex=False)
```

```python
books["imprint"] = books["imprint"].str.replace("O'Reilly Media, Inc.", "O'Reilly", regex=False)
```

```python
books["imprint"] = books["imprint"].str.replace("Barnes Noble Classics", "Barnes & Noble Classics", regex=False)
```
